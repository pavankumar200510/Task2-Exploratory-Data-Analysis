```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/kaggle/input/datasets/pavankumar2910/studentperformance/dataofstudents (1).xlsx")

df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
sns.histplot(df['Exam_Score'], bins=10)
plt.title("Exam Score Distribution")
plt.show()
```

```python
sns.histplot(df['Hours_Studied'])
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.show()
```

```python
sns.scatterplot(x='Attendance', y='Exam_Score', data=df)
plt.show()
```

```python
sns.boxplot(x='Sleep_Hours', y='Exam_Score', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
sns.pairplot(df[['Hours_Studied','Attendance','Sleep_Hours','Exam_Score']])
```

```python
df.groupby('Gender')['Exam_Score'].mean()
```

```python
df.groupby('School_Type')['Exam_Score'].mean()
```

```python
df.groupby('Parental_Education_Level')['Exam_Score'].mean()
```

```python
sns.barplot(x=df.corr(numeric_only=True)['Exam_Score'].sort_values(), 
            y=df.corr(numeric_only=True).index)
```
