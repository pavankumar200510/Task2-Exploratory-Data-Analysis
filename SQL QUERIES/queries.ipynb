```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/kaggle/input/datasets/pavankumar2910/student-performance/dataofstudents (1).xlsx')

df.head()
```

```python
import sqlite3

conn = sqlite3.connect(':memory:')
df.to_sql('students', conn, index=False, if_exists='replace')
```

```python
pd.read_sql("""
SELECT AVG(Exam_Score) AS avg_score
FROM students
""", conn)
```

```python
pd.read_sql("""
SELECT 
    CASE 
        WHEN Hours_Studied < 5 THEN 'Low Study'
        WHEN Hours_Studied BETWEEN 5 AND 10 THEN 'Medium Study'
        ELSE 'High Study'
    END AS Study_Category,
    AVG(Exam_Score) AS avg_score
FROM students
GROUP BY Study_Category;
""", conn)
```

```python
pd.read_sql("""
SELECT 
    CASE 
        WHEN Attendance < 60 THEN 'Low Attendance'
        WHEN Attendance BETWEEN 60 AND 80 THEN 'Medium Attendance'
        ELSE 'High Attendance'
    END AS Attendance_Level,
    AVG(Exam_Score) AS avg_score
FROM students
GROUP BY Attendance_Level;
""", conn)
```

```python
pd.read_sql("""
SELECT Gender, AVG(Exam_Score) AS avg_score
FROM students
GROUP BY Gender;
""", conn)
```

```python
pd.read_sql("""
SELECT MAX(Exam_Score) FROM students;
""", conn)
```

```python
pd.read_sql("""
SELECT MIN(Exam_Score) FROM students;
""", conn)
```

```python
pd.read_sql("""
SELECT AVG(Exam_Score) FROM students;
""", conn)
```

```python
pd.read_sql("""
SELECT COUNT(*) FROM students;
""", conn)
```

```python
pd.read_sql("""
SELECT 
    CASE 
        WHEN Exam_Score >= 50 THEN 'Pass'
        ELSE 'Fail'
    END AS Result,
    COUNT(*) AS total_students
FROM students
GROUP BY Result;
""", conn)
```
